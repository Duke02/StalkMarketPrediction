```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 2, 1, 3, 7])
```

```python
from sklearn.linear_model import LinearRegression
train_x = x[:, np.newaxis]
model = LinearRegression().fit(train_x, y)
y_out = model.predict(train_x)
plt.scatter(x, y)
plt.plot(x, train_x)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3, include_bias=False)
train_x_2 = poly.fit_transform(train_x)
print(train_x_2)

```

```python
model = LinearRegression().fit(train_x_2, y)
y_out = model.predict(train_x_2)
plt.scatter(x, y)
plt.plot(x, y_out)

```
